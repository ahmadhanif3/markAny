```python
import pandas as pd
import os
```

```python
dir_path = '../ransomware_irplog_markany/ransomware_irplog_markany/'
files = [file for file in os.listdir(dir_path) if os.path.isfile(os.path.join(dir_path, file))]

print(len(files))
```

```python
df_pkl = pd.read_pickle('benign/dnn_a2f1b175d19de3bb27895f4dcdcd2fde-001.pkl')
df_pkl.head()
```

```python
major_type = set()
minor_type = set()

for file in files:
    file_path = f'{dir_path}{file}'
    df = pd.read_csv(file_path, sep='\t')
    
    major_type.update(df['Major Operation type'].unique())
    minor_type.update(df['Minor Operation type'].unique())
    
print('Major')
print(major_type)
print('Minor')
print(minor_type)
```
